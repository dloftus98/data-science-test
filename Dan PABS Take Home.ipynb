```python
session_model = pd.read_csv('session_model.csv.gz', compression='gzip', header=0, sep=',', quotechar='"', error_bad_lines=False)
```

```python
session_model
```

```python
session_model['session_ts'] = pd.to_datetime(session_model['timestamp'])

session_model['birth_date'] = pd.to_datetime(session_model['birth_date'], unit='s')
```

```python
session_model = session_model.drop(columns=['timestamp'])
```

```python
session_model.dtypes
```

```python
session_model
```

```python
import datetime
```

```python
session_model['session_date'] = session_model['session_ts'].dt.floor('D')
session_model['session_week'] = session_model['session_ts'].dt.to_period('W')
session_model['session_month'] = session_model['session_ts'].dt.to_period('M')
```

```python
session_model
```

```python
users_per_day = session_model.groupby('session_date')['user_id'].count()
```
